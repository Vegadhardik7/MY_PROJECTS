#### train / fit model into training data

```python
X = data.drop('price',axis=1).values
y= data['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=0.2,random_state=101)
```

#### Feature scaling 

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

#### Creating model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))
```

#### Compile our model

```python
model.compile(optimizer='adam',loss='mse')
```

#### Now Train our model

```python
# After each epochs of training our X_train we will run the test data check our loss on the test data 
# which will help us to keep a track of how well our model is performing not just on our training data but also on our test data

model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,
          epochs=400)
```

```python
loss_val=pd.DataFrame(model.history.history)
```

```python
# loss     =  loss on my training set
# val_loss =  loss on my test set

loss_val
```

```python
loss_val.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score

pred = model.predict(X_test)
pred
```

```python
mean_absolute_error(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
np.sqrt(mean_absolute_error(y_test,pred))
```

```python
data['price'].describe()
```

```python
explained_variance_score(y_test,pred)
```
