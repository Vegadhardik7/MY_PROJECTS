```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"D:\Dig\lending_club_new_data.csv")
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['good_loans']=data['bad_loans'].apply(lambda y:'yes' if y==0 else 'no') 
```

```python
data.head()
```

```python
X=data.drop(['bad_loans','good_loans'],axis=1)
y=data['good_loans']
```

```python
print(X.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Decsison tree model
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
predict = model.predict(X_test)
```

```python
print(confusion_matrix(y_test,predict))
```

```python
#Result
print(classification_report(y_test,predict))
```

```python
#Random Forset
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=250)
rf_model.fit(X_train,y_train)
#Good model but have 16 defects
```

```python
rf_predict=rf_model.predict(X_test)
print(classification_report(y_test,rf_predict))
```

```python
print(confusion_matrix(y_test,rf_predict))
#Better than Decision Tree.......Because it has only 13 defects 
```
