```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"D:\Dig\fake2_data.csv")
data.head()
```

```python
sns.pairplot(data)
```

#### Split Data

```python
from sklearn.model_selection import train_test_split

X = data[['Bedroom','Area']].values
y = data['Price'].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

#### Normalize our Data

```python
from sklearn.preprocessing import MinMaxScaler
# MinMaxScaler is basically transform our data based on the std of data as well as the Min & Max value

scaler = MinMaxScaler()
scaler.fit(X_train)     # fit = Cal the std, min & max
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)

X_train.min()
```

```python
X_train.max()
```

```python
X_train
```

#### Creating a Model

```python
from tensorflow.keras.models import Sequential        #Model
from tensorflow.keras.layers import Dense             #Layer
```

```python
model = Sequential()

model.add(Dense(580,activation='relu'))                 # L 2.1 
model.add(Dense(580,activation='relu'))                 # L 2.2
model.add(Dense(580,activation='relu'))                 # L 2.3

model.add(Dense(1))                                   # L 3
```

#### Compile the model

```python
model.compile(optimizer='rmsprop',loss='mse')
```

#### train / fit model into training data  

```python
model.fit(x = X_train, y = y_train, epochs=2500)
```

```python
loss_df = pd.DataFrame(model.history.history)              # shows the historical losses
```

```python
loss_df.plot()
```

#### Testing / Evaluating our model

```python
model.evaluate(X_test,y_test,verbose=0)      # Returns Models loss on X_test, y_test
```

```python
model.evaluate(X_train,y_train,verbose=0)    # Returns Models loss on X_train, y_train
```

#### Predicting 

```python
test_pred = model.predict(X_test)
test_pred
```

#### Compair true & predicted values

```python
test_pred = pd.Series(test_pred.reshape(4,))
test_pred
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_pred],axis=1)
```

```python
pred_df.columns = ['Test True Y','Model Predictions']
```

```python
pred_df
```

```python
sns.scatterplot(data=pred_df,y='Test True Y',x='Model Predictions')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```
