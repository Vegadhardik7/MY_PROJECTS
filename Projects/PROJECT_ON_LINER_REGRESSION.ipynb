```python
#                DATA ANALYSIS


#Multivariate linear regression

import pandas as pd
import numpy as np
import statsmodels.api as sm
df=pd.read_csv(r"D:\Dig\loan.csv")
df.head()
```

```python
int_rate   = df['Interest_Rate']
loan_amt   = df['Loan_Amount']
fico_score = df['FICO_Score']
```

```python
# transpose = convert rows into columns and columns into rows

y=np.matrix(int_rate).transpose()     #dependent variable
```

```python
x1 = np.matrix(fico_score).transpose()
x2 = np.matrix(loan_amt).transpose()
```

```python
x=np.column_stack([x1,x2])           #dependent varaible
x3 = sm.add_constant(x)
```

```python
#OLS                  ordinary least square  =  estimating the unknown parameters in a linear regression model


model=sm.OLS(y,x3)
model_fit=model.fit()
```

```python
print("P values are : ",model_fit.pvalues )
print("The R square value is : ",model_fit.rsquared)  # rsquared is called as co-efficient of determination
                                                      # how well the regression line approximate the real data                  
```
