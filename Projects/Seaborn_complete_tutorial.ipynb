## How To Draw seaborn Histogram / seaborn Distplot

```python
#distplot use to plot histogram
sns.distplot(tips_df["size"])
#the curve line is known as KDE = Kernel Density Estimator
```

```python
sns.distplot(tips_df["tip"])
```

```python
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"])
```

```python
#Parameters in histogram
# sns.distplot(
#     a,
#     bins=None,
#     hist=True,
#     kde=True,
#     rug=False,
#     fit=None,
#     hist_kws=None,
#     kde_kws=None,
#     rug_kws=None,
#     fit_kws=None,
#     color=None,
#     vertical=False,
#     norm_hist=False,
#     axlabel=None,
#     label=None,
#     ax=None,
# )
```

```python
#divide our data in n number of bars
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],bins=100)
```

```python
#if you only want the curve line
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],hist=False)
```

```python
#if you only want the histogram
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],kde=False)
```

```python
#show exact value of data
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],rug=True)
```

```python
#if you only want normalized curve line
from scipy.stats import norm
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],fit=norm)
```

```python
#if you want to change color
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],color="red")
```

```python
#if you only want histogram to be horizontal
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],vertical=True)
```

```python
#if you only want to change the x axis name
plt.figure(figsize=(20,9))
sns.distplot(tips_df["total_bill"],axlabel="Total Bill")
```

```python
plt.figure(figsize=(20,9))
sns.set(style='darkgrid')
sns.distplot(tips_df["total_bill"],label="Bill")
plt.title("Histogram of Bill")
plt.legend()
```

```python
#want to show values in a perticular range
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(20,9))
sns.set(style='darkgrid')
sns.distplot(tips_df["total_bill"],bins=bins)
plt.xticks(bins)
plt.title("Histogram of Bill")
plt.legend()
```

```python
plt.figure(figsize=(20,9))
sns.set(style='darkgrid')
sns.distplot(tips_df["total_bill"],hist_kws={'color':'red','edgecolor':'yellow','linewidth':5,'linestyle':'--','alpha':0.9})
```

```python
plt.figure(figsize=(20,9))
sns.set(style='darkgrid')
sns.distplot(tips_df["total_bill"],hist_kws={'color':'red','edgecolor':'yellow','linewidth':5,'linestyle':'--','alpha':0.9}
                                  ,kde_kws={'color':'BLUE','linewidth':5,'linestyle':'--','alpha':0.9}
                                  ,rug=True,
                                   rug_kws={'color':'white','edgecolor':'black','linewidth':1,'linestyle':'--','alpha':0.9})
```

```python
plt.figure(figsize=(20,9))
sns.set(style='darkgrid')
sns.distplot(tips_df["total_bill"],hist_kws={'color':'red','edgecolor':'yellow','linewidth':5,'linestyle':'--','alpha':0.9}
                                  ,kde=False,
                                   fit=norm,
                                   fit_kws={'color':'BLUE','linewidth':8,'linestyle':'--','alpha':0.5}
                                  ,rug=True,
                                   rug_kws={'color':'white','edgecolor':'black','linewidth':1,'linestyle':'--','alpha':0.9})
```

```python
#plot multipal histogram in one figure
plt.figure(figsize=(20,10))
sns.distplot(tips_df["tip"],bins=1,label="tip")

sns.distplot(tips_df["total_bill"],label="total Bill",bins=1,
             hist_kws={'color':'red','edgecolor':'yellow','linewidth':5,'linestyle':'--','alpha':0.9})

sns.distplot(tips_df["size"],bins=1,label="size",rug=True,
             hist_kws={'color':'blue','edgecolor':'white','linewidth':5,'linestyle':'--','alpha':0.9})

plt.legend()
```

```python
sns.set()
plt.figure(figsize=(15,5))
kwargs={'alpha':0.96,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,n_boot=2, **kwargs)  #**kwargs means passing as dictionary
```

```python
sns.set()
plt.figure(figsize=(15,5))
kwargs={'alpha':0.96,'linestyle':':','linewidth':5,'edgecolor':'k'}
ax=sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,n_boot=2, **kwargs)
ax.set(title="Bar plot bill",
      xlabel="Days",
      ylabel="Total Bill")
```

```python
sns.set()
plt.figure(figsize=(15,5))
kwargs={'alpha':0.96,'linestyle':':','linewidth':5,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,n_boot=2, **kwargs)
plt.title("Bar days & total bill",fontsize=20)
plt.xlabel("Days",fontsize=20)
plt.ylabel("Bill",fontsize=20)
plt.savefig("Bar days & total bill")
plt.show()
```
