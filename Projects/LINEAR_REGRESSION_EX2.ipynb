```python
data['Engine_HP'] = data['Engine_HP'].fillna((data['Engine_HP'].mean()))
```

```python
data['Engine_Cylinders'] = data['Engine_Cylinders'].fillna((data['Engine_Cylinders'].mean()))
```

```python
data['Number_of_Doors'] = data['Number_of_Doors'].fillna((data['Number_of_Doors'].mean()))
```

```python
data.info()
```

```python
#Year Engine_HP Engine_Cylinders Number_of_Doors highway_MPG city_mpg Popularity Price

from sklearn.model_selection import train_test_split


X=data[['Year' , 'Engine_HP' , 'Engine_Cylinders' , 'Number_of_Doors' , 'highway_MPG' , 'city_mpg' , 'Popularity']]
y=data['Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
pred[1]
```

```python
model.score(X,y)
```

```python
#caluclate root mean square error
np.mean((y_test - pred)**2)
```

```python
from sklearn.metrics import mean_squared_error
mean_sq_err = mean_squared_error(y_test,pred)
```

```python
mean_sq_err
```

```python
RMSE = np.sqrt(mean_sq_err)
```
