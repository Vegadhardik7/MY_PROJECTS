# Question 1 – Replace Vowels with a digit

```python
import math
from math import sqrt 

vowels = ['a','e','i','o','u']

class GenerateCode:                              
    def __init__(self, data):
        self.data = data
    
    
    # Step 1: Get the index of the vowel in the given input
    def get_vowel_index(self):
        res = []                                # Empty List to append all the index values which are present in string
        for i in range(len(self.data)):         # Iterating through all the values of the string 
            if self.data[i] in vowels:          # Trying to match value with vowels 
                res.append(i)                   # If matched than append that value in the empty list
        return res                              # Finally returning the list
    
    
    
    
    # Step 2: Multiply the index with 100
    def mul_by_100(self,res):                   
        values_mul_100 = []                     # Empty List to append all the values from res which will be multiplied with 100
        for i in res:                           # For all the element in res
            value = i * 100                     # Multiply each value with 100
            values_mul_100.append(value)        # Appending all the value which we got into values_mul_100
        return values_mul_100                   # Finally returning the list
    
    
    

    # Step 3: Sum all the prime numbers between 1 and the resulting number from last step.
    def sum_of_prime(self, value):
        sumval = 0                              # Taking inital variable to increment
        for x in range(1,value+1):              # Iterating through 1 to Nth element + 1, because it consider N-1
            if x == 1:                          # If value of x is 1 then pass
                pass 
            else:                               
                half = int(sqrt(x)) + 1         # By default we will get value as float so converting it to int
                for y in range(2,half):         # Iterating though the range of 2 till the value "half" to get the result
                    res = x%y                   # Taking the mod of variable x and y (HELP TO GET REMAINDER of x/y)
                    if res==0:                  # If our reslut is 0 then break
                        break
                else:                           # Else increment the value of sum with value if x
                    sumval = sumval + x           
        return sumval                           # Return sumval variable final value (Give us prime number)

    
    
    
    # Step 4: Add the digits of the number you get from the last step until you get a single digit.
    def sum_of_digits(self, num):
        return (num - 1) % 9 + 1 if num > 0 else 0 
    
    
    

    # Step 5: Now, Replace the vowel with the digit from the last step.
    def replace_vowels(self, data, making_single):
        i = 0 
        for k in self.data:                                             # For all the element in our string 
            if k.lower() in vowels:                                     # If that element is present in list of vowel
                self.data = self.data.replace(k,str(making_single[i]),1)# Then replace it with the making_string's each var  
                i += 1                                                  # Then increment it by 1
            if i>=5:                                                    # If the value excits the range than 0
                i=0
        return self.data                                                # returning our final value
    
    
# Initializing a string to a variable
input_string = 'replace this'

# Making an Class object with string variable initialize in it
code = GenerateCode(input_string)

# Getting the vowel index values in a list
vowels_index = code.get_vowel_index()
print(f"Getting Vowels Index List:\n\n {vowels_index}\n\n")


# Multiplying all the values with 100 and getting the result
vowels_index100 = code.mul_by_100(vowels_index)
print(f"Multiplying all values in {vowels_index} with 100:\n\n {vowels_index100}\n\n")


# Sum all the prime numbers between 1 and the resulting number from last step
lst_of_sum_of_prime = list(map(code.sum_of_prime,vowels_index100))
print(f"Sum all the prime numbers between 1 and the resulting number from last step:\n\n {lst_of_sum_of_prime}\n\n")


# Add the digits of the number you get from the last step until you get a single digit
making_single = list(map(code.sum_of_digits, lst_of_sum_of_prime))
print(f"Converting all the code into single digit number:\n\n {making_single}\n\n")


# Now, Replace the vowel with the digit from the last step
final = code.replace_vowels(input_string,making_single)
print(f"Repalcing all vowels with {making_single}:\n\nFINAL OUTPUT:\n\n {final}")
```

# Question 2 – Minimum Swaps to Symmetric String

```python
def ValueSwap(s, n):
    s = list(s)
 
    # Counter to count minimum swap
    count = 0
    ans = True
 
    # A loop which run in half string
    # from starting
    for i in range(n // 2):
 
        # Left pointer
        left = i
 
        # Right pointer
        right = n - left - 1
 
        # A loop which run from right pointer
        # to left pointer
        while left < right:
 
            # if both char same then
            # break the loop if not
            # same then we have to move
            # right pointer to one step left
            if s[left] == s[right]:
                break
            else:
                right -= 1
 
        # it denotes both pointer at
        # same position and we don't
        # have sufficient char to make
        # palindrome string
        if left == right:
            ans = False
            break
        else:
            for j in range(right, n - left - 1):
                (s[j], s[j + 1]) = (s[j + 1], s[j])
                count += 1
    if ans:
        return (count)
    else:
        return -1
 
 
# Driver Code
s = 'abca'
 
# Legth of string
n = len(s)
 
# Function calling
ans1 = ValueSwap(s, n)
ans2 = ValueSwap(s[::-1], n)
print(max(ans1, ans2))
```
